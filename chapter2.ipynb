```python
# r18 -
from abc import ABCMeta, abstractmethod
class progression(metaclass = ABCMeta):
    
    def __init__(self, start= 0):
        self._current = start
        
    def __iter__(self):
        return self
    @abstractmethod
    def print_progression(self,n):
        pass

    def __next__(self):
        ans = self._current
        self._advance()
        return ans

    
    @abstractmethod
    def _advance():
        pass
        
        
    @abstractmethod
    def nth(self, n):
        pass
        
    
    
        
```

```python
#r 18
class fibonacci_progression(progression):
    def __init__(self,first, second):
        self.first = first
        self.second = second
        super().__init__(first)
        self._prev = second - first

    def _advance(self):
        self._prev,self._current = self._current, self._prev+self._current

    def print_progression(self, n):
        self.__init__(self.first, self.second)
        return ",".join(str((next(self))) for _ in range(n))

    def nth(self, n):
        val = self.print_progression(n)
        return val[val.rfind(',')+1:] # use msm method to find nth number of fibonacci.
        
```
